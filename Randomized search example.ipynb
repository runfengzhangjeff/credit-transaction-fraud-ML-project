参照自己model的可以调的参数来改distribution部分的dict。每个参数都规定一个distribution当做范围。要注意有些参数只能integer比如max_depth这种就用randint，continous的参数就用uniform

```python
from scipy.stats import uniform
from scipy.stats import randint
from sklearn.model_selection import RandomizedSearchCV
xgbo = xgb.XGBRegressor()
distributions = dict(n_estimators=randint(low=50, high=350),
                     max_depth = randint(low=1, high=5),
                     learning_rate=uniform(loc=0, scale=0.5),
                     min_child_weight = [ 1, 3, 5, 7 ],
                     gamma           = uniform(loc=0, scale=0.5),
                     reg_alpha            = uniform(loc=0, scale=1),
                     reg_lambda            = uniform(loc=0, scale=1),
                    booster=['gbtree', 'gblinear' ,'dart'])
clf = RandomizedSearchCV(xgbo, distributions, random_state=0)
search = clf.fit(new_x, y)
a=search.best_params_
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
kfolds_regresssion = KFold(n_splits = 10, random_state = 1, shuffle = True)
regresssion_model = xgb.XGBRegressor(params=a)
neg_mse_model_2_cv = cross_val_score(regresssion_model,new_x, y,cv=kfolds_regresssion,scoring = 'r2')
print("XGB Regression: \n")
print("r squared of 10-folds:",neg_mse_model_2_cv,"(mean r squared:",np.mean(neg_mse_model_2_cv),")")
```
